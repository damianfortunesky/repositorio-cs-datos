```
# imports y datos
import pandas as pd
df = pd.read_csv('data/ventas.csv')
df
```

```
# EXPLORO LAS COLUMNAS DEL DF
print(df.columns)
```

```
# cuantos registros tiene el dataframe?
print('Total de operaciones:', df['Importe'].count())
```

```
# qué tipo de datos tiene?
df.dtypes
```

```
#CONVIERTO LA COLUMNAS DE importes a FLOAT
df['Importe'] = df['Importe'].astype(float)
df['Comision_x_venta'] = df['Comision_x_venta'].astype(float)
df.dtypes
```

```
# VERIFICO
# qué tipo de datos tiene ahora?

df.dtypes
```

```
# LE DOY FORMATO DE $$$$ (sólo la apariencia al imprimir)
pd.options.display.float_format = '$ {:,.2f}'.format # corre para TODAS las columnas float
df['Importe']
```

```
#ORDENAR EL DATAFRAME POR NOMBRE DEL VENDEDOR
print(df.sort_values('Vendedor'))
# tarea: investigar ordenamiento descendente
```

```
#OBTENER EL MÁXIMO DE TODAS LAS COLUMNAS
print(df.max())
```

```
# ACTIVIDAD DEL VENDEDOR Mufi Panchin con boolean masking
print(df.Vendedor == 'Mufi Panchin')
```

```
# obtengo los registros completos de actividad de Mufi Panchin
mufi = df.loc[df['Vendedor'] == 'Mufi Panchin']
mufi
```

```
# BUSCO LOS REGISTROS DONDE EL VENDEDOR SEA "Mufi Panchin" y Sede sea "Paternal"
print(df.loc[(df['Vendedor'] == 'Mufi Panchin') & (df['Sede']=='Paternal')])
```

```
# CARGO LAS MEJORES VENTAS (>200000) EN OTRO DATAFRAME
mejores = df.loc[df['Importe'] > 200000]

# Y BUSCO A MUFI PANCHIN en el nuevo df
mmufi = mejores.loc[mejores['Vendedor'] == 'Mufi Panchin']
print(mmufi)
```

```
# cuántas operaciones >200000 concretó Mufi Panchin?
print(mejores['Importe'].count())
print(mmufi['Importe'].count())
```

```
# QUIERO VER QUÉ RUBROS VENDIÓ MUFI PANCHIN Y POR QUÉ IMPORTE
#print(mufi['Rubro'],mufi['Importe']) # fallido1
#print(mufi['Rubro','Importe']) # fallido2
```

```
print(mufi[['Rubro'] + ['Importe']]) # OJO: dentro de unpar de corchetes con +
```

```
# MALA PRÁCTICA N° 1: Modificar el DF para filtrar --> debo extraer a otro df
# si el importe es menor a 350000, cambiarlo a None(ninguno)
mufi.loc[mufi['Importe'] < 350000, 'Importe'] = None
print(mufi[['Rubro'] + ['Importe']])
# NaN (not a number) / None / Null
```

```
# cuando haga un count, Los None van a ser ignorados
print(mufi['Importe'].count())
```

```
# MALA PRÁCTICA N° 2: Inventar los datos
mufi = mufi.fillna(0)
print(mufi[['Rubro'] + ['Importe']])
print(mufi['Importe'].count())
```
