<h1>Seaborn</h1>
<center><img src='/work/img/sns_logo.png'></center>
<p>Es una librería construida sobre Matplotlib, optimizada para trabajar con dataframes pandas.</p>
<h4>Estructura básica: </h4>
<code>
sns.'Tipo de Grafica'(<br>
data='Dataset',<br>
x='Data en el eje x',<br>
y='Data en el eje y',<br>
hue='Variable de agrupamiento')<br>
</code>
<h4>Tipos de gráficas:</h4>
<ul>
<li><b>relplot (relacional):</b> scatterplot, lineplot.</li>
<li><b>displot (distribuciones):</b> histplot, kdeplot, ecdfplt, rugplot.</li> 
<li><b>catplot (categórica):</b> stripplot, swarmplot, boxplot, violinplot, pointplot, barplot.</li>  
</ul>
<h2>Barplot</h2>


```
# imports y datos
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

sns.set(style='dark', palette='dark')

```

<h2>Stripplot y swarmplot

```
sns.stripplot(data=tips, x='day', y='total_bill', hue='smoker')
```

```
# dodge
sns.stripplot(data=tips, x='day', y='total_bill', hue='smoker', dodge=True)
```

```
sns.swarmplot(data=tips, x='day', y='total_bill', hue='smoker', dodge=True)
```

<h2>Gráficas combinadas</h2>

```
sns.boxplot(data=tips, x='day', y='total_bill', hue='smoker', dodge=True)
sns.swarmplot(data=tips, x='day', y='total_bill', hue='smoker', dodge=True, color='yellow', marker='*')
```

<h2>Violín</h2>
<p>Es una gráfica parecida a boxplot, pero en lugar de cuartiles, muestra la concentración de datos. Admite función <b>split</b> para comparar variables categóricas.</p>

```
sns.violinplot(data=tips, x='day', y='total_bill', hue='smoker')
```

```
#split 
sns.violinplot(data=tips, x='day', y='total_bill', hue='smoker', split=True)
```

<h2>Catplot</h2>
<p>Podemos combinar con otra variable usando <b>col</b> y seleccionar el tipo de gráfica con kind: </p>
<ul>
<li>strip (default)</li>
<li>swarm</li>
<li>box</li>
<li>violin</li>
<li>boxen</li>
<li>point</li>
<li>bar</li>
<li>count</li>
<ul>


```
sns.catplot(data=tips, x='day', y='total_bill', hue='smoker', dodge=True, col='time', kind='violin')
```

<h2>Pairplot</h2>
<p>Correalciones todas las variables numéricas entre sí.</p>
<p>Podemos eliminar las gráficas repetidas.</p>

```
sns.pairplot(data=tips)
```

```
# Eliminamos gráficas repetidas
sns.pairplot(data=tips, corner = True)
```
