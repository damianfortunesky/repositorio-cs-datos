```
# imports y datos
import matplotlib.pyplot as plt
import numpy as np

arts = ['PANTALÃ“N', 'REMERA', 'BLUSA', 'CAMISA', 'POLLERA', 'ABRIGO']
stock = [1524, 2684, 800, 1756, 1008, 3521]

plt.bar(arts, stock)
plt.show()
```

```
# COLORES Y TAMAÃ‘O
plt.bar(arts, stock, width=0.5, color=['blue', 'plum', 'teal', 'hotpink', 'aqua', 'pink'])
plt.show()
```

```
# cambiar nombre de las variables en el eje x con xticks y rotation
plt.bar(arts, stock, width=0.5, color=['blue', 'plum', 'teal', 'hotpink', 'aqua', 'pink'])
plt.xticks(np.arange(6),('PantalÃ³n','Remera','Blusa','Camisa','Pollera','Abrigo'))
plt.show()
```

```
# barh
plt.barh(arts, stock)
plt.show()
```
