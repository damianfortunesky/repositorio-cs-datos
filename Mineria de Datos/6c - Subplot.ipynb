<h2>Subplot</h2>
<p>Permite crear gráficos dentro de una sola. Esto lo hace a través de una <b>matriz de gráficos</b> y se puede acceder a cada uno a través de sus índices.</p>


```
# imports y datos
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2

x
```

```
# SINTAXIS: matplotlib.pyplot.subplot(filas,columnas,index)
# gráfico 1
plt.subplot(2,2,1) # plot 1
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')

# gráfico 2
plt.subplot(2,2,2) # plot 2
plt.pie(y) 

# gráfico 3
plt.subplot(2,2,4) 
plt.hist(y)

# gráfico 4
plt.subplot(2,2,3) 
plt.pie(x)

plt.show()
```

```
# Invertir la matriz de gráficos
# gráfico 1
plt.subplot(1,4,1) # plot 1
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')

# gráfico 2
plt.subplot(1,4,2) # plot 2
plt.pie(y) 

# gráfico 3
plt.subplot(1,4,4) 
plt.hist(y)

# gráfico 4
plt.subplot(1,4,3) 
plt.pie(x)

plt.show()
```

```
# gráfico 1
plt.subplot(4,1,1) # plot 1
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')

# gráfico 2
plt.subplot(4,1,2) # plot 2
plt.pie(y) 

# gráfico 3
plt.subplot(4,1,4) 
plt.hist(y)

# gráfico 4
plt.subplot(4,1,3) 
plt.pie(x)

plt.show()
```
