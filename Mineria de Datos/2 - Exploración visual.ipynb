<h3>Seaborn</h3>
<p>Seaborn es una librería de visualización de datos para Python basada en Matplotlib.</p>
<p>Además de las funciones de visualización, nos provee de diversos datasets para práctica, a los que podemos acceder con el método <code><mark>seaborn.load_dataset()</code></mark>. Podemos encontrar la lista de datasets disponibles en <a href="https://github.com/mwaskom/seaborn-data">https://github.com/mwaskom/seaborn-data</a></p>

```
# Documentación: https://seaborn.pydata.org/
import pandas as pd
import seaborn as sns
dftips = sns.load_dataset('tips')
dftips
```

```
# DIAGRAMA DE DISTRIBUCIÓN
sns.displot(dftips.tip)
```

```
# DEPRECATED --> OBSOLETO
sns.distplot(dftips.tip)
```

```
sns.kdeplot(dftips.tip, fill = True)
# shade = True deprecated
```

```
tM = dftips[dftips['sex']=='Male']
tF = dftips[dftips['sex']=='Female'].copy()
sns.kdeplot(tM.tip, fill = True, color='blue', alpha=0.6)
sns.kdeplot(tF.tip, fill = True, color='red', alpha=0.2)
```

```
sns.histplot(tM.tip, color='blue')
sns.histplot(tF.tip, color='red')
```

```
# Diagramas de dispersión: scatterplot 
sns.scatterplot(x='total_bill', y='tip', data=dftips)
```

```
# Podemos discriminar por variables categóricas 
sns.scatterplot(x='total_bill', y='tip', data=dftips, hue='sex')
```

```
sns.scatterplot(x='total_bill', y='tip', data=dftips, hue='sex', style='time')
```

```
# Relplot
sns.relplot(x='total_bill', y='tip', data=dftips, hue='sex')
```

```
# joint plot
# permite hacer diagrama de dispersion combinado con el histo de distribución

sns.jointplot(x='total_bill', y='tip', data=dftips)
```

```
sns.jointplot(x='total_bill', y='tip', data=dftips, hue='smoker')
```

<h4>Graficando variables categóricas

```
sns.relplot(x='total_bill', y='tip', hue='time', col='smoker', data=dftips)
```

```
sns.relplot(x='total_bill', y='tip', hue='time', col='sex', data=dftips)
```

```
sns.relplot(x='total_bill', y='tip', hue='time', col='day', data=dftips)
```

```
sns.catplot(x='day',y='total_bill', hue='sex', data=dftips)
```

```
sns.catplot(x='day',y='total_bill', hue='sex', data=dftips, alpha=0.43)
```

```
sns.boxplot(y='day', x='total_bill',data=dftips)
```

```
sns.boxplot(y='sex', x='total_bill',data=dftips)
```

```
sns.boxplot(x='day', y='total_bill', hue='sex', data=dftips)
```

```
#DEPURANDO LOS DATOS -> Me interesan sólo fines de semana
dftips['weekend'] = dftips['day'].isin(['Sat','Sun'])
dftips

```

```
sns.boxplot(y='day', x='total_bill', hue='weekend', data=dftips)
```

```
sns.barplot(x='sex', y='total_bill', hue='smoker', data=dftips)
```

```
sns.barplot(x='day', y='tip',hue='smoker',palette='ch:.25', data=dftips)
```

```
# CONTEO en lugar de usar una columna como métrica
sns.countplot(x='day', hue='time',data=dftips)
```

```
# CATPLOT: muestra la relación entre una variable numérica y una o más variables categóricas
sns.catplot(x='smoker', y='total_bill',col='sex',hue='day',kind='bar',data=dftips)
```

```
# "Más" no siempre es mejor. 
# necesito hacer una pregunta para saber lo que estoy buscando. 
# Quitando la variable categórica 'day' se ve a simple vista la diferencia de comportamiento 

sns.catplot(x='smoker', y='total_bill',col='sex',kind='bar',data=dftips)
```
