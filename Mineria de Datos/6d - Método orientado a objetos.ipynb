<h2>Método orientado a objetos</h2>
<p>Documentación: <a href ='https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html'>https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html</a>
<p>Escribiendo un poco más de código podemos facilitar el proceso de personalización de nuestras gráficas.</p>
<p>Empecemos por conocer la estructura de una figura:</p>
<center><img src='/work/img/figure.png'></center>
<ul>
<li>Figure: es el lienzo donde voy a ubicar mis gráficas</li>
<li>Axes: son las gráficas individuales
</li>
</ul>

```
# import y datos
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,20)
y = x ** 2
```

```
# creo la figura = lienzo
fig = plt.figure()

# agrego uyna gráfica (axes) al lienzo
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9]) # ---> [posx, posy, sizex, sizey]

# especifico la gráfica
axes.plot(x,y,'b')

fig.show() # dispara la gráfica
```

```
# creo la figura = lienzo
fig = plt.figure()

# agrego una gráfica (axes) al lienzo
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9]) # ---> [posx, posy, sizex, sizey]

# agrego otra gráfica (axes) al lienzo
axes2 = fig.add_axes([0.2, 0.6, 0.2, 0.3]) # ---> [posx, posy, sizex, sizey]

# especifico la gráfica
axes.plot(x,y,'b')
axes2.plot(y,x,'r')

axes2.set_facecolor('lightcyan')

fig.show() # dispara la gráfica
```

<p>Ambos axes son diferentes objetos dentro del mismo lienzo.</p>
<h2>Subplots</h2>
<p>Permite trabajar con varias gráficas a la vez, aprovechando las facilidades de objetos.</p>

```
x = np.linspace(0,7,11)
y = np.sin(x)
```

```
fig, axes = plt.subplots() # ojo: este es con 's' final
axes.plot(x,y,'r:')
plt.show()
```

```
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'r:')
axes[1].plot(y,x,'b-')
plt.show()
```

```
#acceder a los gráficos mediante el nombre almacenado en una tupla
fig, (axes1, axes2) = plt.subplots(nrows=1, ncols=2)
axes1.plot(x,y,'r:')
axes2.plot(y,x,'b-')
plt.show()
```

```
#Generar una matriz de tipo fig, con dos filas y cuatro columnas
v = np.cos(x)
w = x**2
z = np.tan(x)

fig, axes = plt.subplots(2,4) # si respeto el orden toma fila, columna
axes[0,0].plot(x,y,'g')
axes[0,1].plot(y,x,'r')
axes[0,2].plot(z,y, 'magenta')
axes[0,3].plot(y,z, 'blue')

axes[1,0].plot(x,v, 'y')
axes[1,1].plot(v,x, 'turquoise')
axes[1,2].plot(x,w, 'hotpink')
axes[1,3].plot(w,x, 'coral')
 
fig.tight_layout() # mejora la visualización de los ejes de cada gráfico
```

```
#Alternativa para generar una matriz de tipo fig, accediendo a cada gráfica mediante los nombres dados en una tupla
fig, ((a1, a2, a3, a4),(a5, a6, a7, a8)) = plt.subplots(2,4)

a1.plot(x, np.sin(x), 'b')
a2.plot(y, np.cos(x), 'purple')
a3.plot(x, np.tan(y), 'orange')
a4.plot(y, np.cos(y)**2, 'black')

a5.plot(x, np.negative(np.sin(x)), 'y')
a6.plot(y, np.negative(np.cos(x)), 'hotpink')
a7.plot(x, np.negative(np.tan(y)), 'brown')
a8.plot(y, np.negative(np.cos(y)), 'green')

fig.tight_layout()

```
