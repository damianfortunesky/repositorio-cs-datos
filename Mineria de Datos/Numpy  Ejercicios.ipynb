# Numpy Ejercicios

## Ejercicio 1

Crear un vector con valores dentro del rango 10 a 49.

## Ejercicio 2

Invertir el vector.

## Ejercicio 3

Crear una matriz 3x3 con valores de 0 a 8.

## Ejercicio 4

Encontrar los indices que no son ceros del arreglo [1,2,4,2,4,0,1,0,0,0,12,4,5,6,7,0]

## Ejercicio 5

Crear una matriz identidad de 6x6.

## Ejercicio 6

Crear una matriz con valores al azar con forma 3x3x3.

## Ejercicio 7

Encontrar los indices de los valores minimos y maximos de la anterior matriz

## Ejercicio 8

Crear una matriz de 10x10 con 1's en los bordes y 0 en el interior (con rangos de indices)

## Ejercicio 9

Crear una matriz de 5x5 con valores en los renglones que vayan de 0 a 4

## Ejercicio 10

Crear dos arreglos al azar A y B, verificar si son iguales

```
import numpy as np

# Ejercicio 1: Crear un vector con valores dentro del rango 10 a 49
vector = np.arange(10, 50)
print("Ejercicio 1:", vector)

# Ejercicio 2: Invertir el vector
vector_invertido = vector[::-1]
print("Ejercicio 2:", vector_invertido)

# Ejercicio 3: Crear una matriz 3x3 con valores de 0 a 8
matriz_3x3 = np.arange(9).reshape(3, 3)
print("Ejercicio 3:\n", matriz_3x3)

# Ejercicio 4: Encontrar los indices que no son ceros del arreglo
arreglo = np.array([1, 2, 4, 2, 4, 0, 1, 0, 0, 0, 12, 4, 5, 6, 7, 0])
arreglo_sin_ceros = np.nonzero(arreglo)
print("Ejercicio 4:", arreglo_sin_ceros)

# Ejercicio 5: Crear una matriz identidad de 6x6
matriz_identidad = np.eye(6)
print("Ejercicio 5:\n", matriz_identidad)

# Ejercicio 6: Crear una matriz con valores al azar con forma 3x3x3
matriz_azar = np.random.random((3, 3, 3))
print("Ejercicio 6:\n", matriz_azar)

# Ejercicio 7: Encontrar los indices de los valores minimos y maximos de la matriz anterior
indice_min = np.unravel_index(np.argmin(matriz_azar), matriz_azar.shape)
indice_max = np.unravel_index(np.argmax(matriz_azar), matriz_azar.shape)
print("Ejercicio 7: Mínimo en", indice_min, ", Máximo en", indice_max)

# Ejercicio 8: Crear una matriz de 10x10 con 1's en los bordes y 0 en el interior
matriz_10x10 = np.ones((10, 10))
matriz_10x10[1:-1, 1:-1] = 0
print("Ejercicio 8:\n", matriz_10x10)

# Ejercicio 9: Crear una matriz de 5x5 con valores en los renglones que vayan de 0 a 4
matriz_5x5 = np.tile(np.arange(5), (5, 1))
print("Ejercicio 9:\n", matriz_5x5)

# Ejercicio 10: Crear dos arreglos al azar A y B, verificar si son iguales
A = np.random.random(10)
B = np.random.random(10)
iguales = np.array_equal(A, B)
print("Ejercicio 10: ¿Son iguales A y B?", iguales)
```
