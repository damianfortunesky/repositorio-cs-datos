<h2>Contexto</h2>
<p>El contexto facilita la lectura de nuestras gráficas: etiquetas, leyendas y títulos</p>

```
# imports y datos
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,20)
y = x**2
```

```
fig, (a1, a2) = plt.subplots(1,2)

a1.plot(x,y, 'g', label = 'Creciente')  # este label es para legend
a2.plot(y,x, 'r', label = 'Decreciente')

# Títulos
a1.set_title('Relación x/y')
a2.set_title('Relación y/x')

# Labels = nombre de los ejes
a1.set_xlabel('Valor x')
a1.set_ylabel('Valor y')

a2.set_xlabel('Valor y')
a2.set_ylabel('Valor x')

# Legend 
# localizaciones: 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'upper center'...
a1.legend() # activar el label de ax.plot
a2.legend(loc='lower right')

plt.show()
```

<h2>Estilos</h2>

```
print('¿Qué estilos hay disponibles en matplotlib.pyplot?')
# https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
print(plt.style.available)
```

```
# UTILIZAR ESTILOS
plt.plot(x,y)
plt.show()
```

```
plt.style.use('seaborn-v0_8-whitegrid')
plt.plot(x,y)
plt.show()
```

```
plt.style.use('Solarize_Light2')
plt.plot(x,y)
plt.show()
```

```
# GROSOR DE LÍNEA (linewidth)
# TRANSPARENCIA (alpha)
plt.plot(x,y, linewidth=7, alpha=0.5)
plt.show()
```

```
# ESTILO DE LÍNEA : linestyle
# MARKERS : marker y markersize
plt.style.use('dark_background')
plt.plot(x,y,'y', linestyle='dashed', marker='*', linewidth=7, alpha=0.5)
plt.show()
```
