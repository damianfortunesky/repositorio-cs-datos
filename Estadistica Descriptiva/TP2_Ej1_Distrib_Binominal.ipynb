#**Parámetros de distribución binominal**


```python
#Número de preguntas
n=10

#Probabilidad de acertar una pregunta
p = 0.25

#Maximos y minimos
min = 0

max = n
```

Defino las variables x = n de preguntas acertadas, y = n preguntas erradas
X es una variable aleatoria discreta con valores  0,1,2,...,10.
X tiene distribucion binominal con parametros n y p, x~BI (n,p) que cumple 3 condiciones:

1) Cada ensayo (responder una pregunta del examen) tiene la misma probabilidad de exito.  

2) El resultado de cada ensayo (si acierta o errra la pregunta) no influye en los resultados de los otros, es decir son independientes.

3) La variable x es igual al número de éxitos en n ensayos.


Creamos un dataframe llamado “binomial” con una columna “X” que contiene valores del 0 al 10, y otra columna “PX” que representa la probabilidad puntual de cada valor de X.

```python
binomial = pd.DataFrame()
binomial['X'] = range(min, max+1, 1)
binomial['px'] = [binom.pmf (k = r , n = n , p = p ) for r in binomial['X']]
print(binomial)
```
