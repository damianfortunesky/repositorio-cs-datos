# **Importación de Recursos**


```python
import pandas as pd
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
import seaborn as sns
```

#**Defino la variable X (éxito)**

X = cantidad de lanzamientos a un tablero hasta acertar con la zona central



#**Probabilidad de éxito**

p=0.17

```python
# Parámetro de la distribución geométrica
p = 0.17

# Definir el rango de valores para la variable aleatoria
x = np.arange(1, 31)

# Calcular las probabilidades de la distribución geométrica
pmf = geom.pmf(x, p)

# Crear la gráfica de barras
plt.figure(figsize=(10, 6))
sns.barplot(x=x, y=pmf, color='blue')

# Añadir etiquetas y título
plt.xlabel('Número de ensayos hasta el primer éxito')
plt.ylabel('Probabilidad')
plt.title('Distribución Geométrica (p = 0.17)')

# Mostrar la gráfica
plt.show()
```

**Respuesta al ejercicio**

```python
probabilidad = geom.pmf(8, p)
print(f"La probabilidad de que se necesiten exactamente 8 lanzamientos para acertar al centro es: {probabilidad:.4f}")
```

#**Media y Varianza**

```python
# Calcular la media y la varianza
media = 1 / p
varianza = (1 - p) / p**2

print(f"Media (esperanza): {media}")
print(f"Varianza: {varianza}")
```
