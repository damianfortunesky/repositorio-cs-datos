```python
import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
def dist_normal(media, var):
  normal = pd.DataFrame()
  normal['X'] = np.arange(media-3*np.sqrt(var), media+3*np.sqrt(var), 0.001)
  normal['fx'] = norm.pdf (normal['X'], media, np.sqrt(var))
  fig = plt.plot (normal['X'], normal['fx'])
  print(f'La función de densidad de una variable que se distribuye como normal con media {media} y varianza {var} es:')
  print(normal,'\n')
  print(f'La gráfico de la función de densidad de una variable que se distribuye como normal con media {media} y varianza {var} es:')
  plt.show(fig)
```

**Observación**: El ejercicio brindaba el dato del desvío estandar. Simplemente, se elevó al cuadrado para obtener la varianza. Siendo la media=30 y la varianza = 4 entonces X tiene distribución N-(30,4)

```python
dist_normal(30, 4)
```
